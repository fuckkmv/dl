```python
from tensorflow import keras as kr
```

```python
(X_train, y_train), (X_test, y_test) = kr.datasets.boston_housing.load_data()
```

```python
model = kr.models.Sequential()

model.add(kr.layers.Dense(X_train.shape[1], activation=kr.activations.relu))
model.add(kr.layers.Dense(50, activation=kr.activations.relu))
model.add(kr.layers.Dense(100, activation=kr.activations.relu))
model.add(kr.layers.Dense(50, activation=kr.activations.relu))
model.add(kr.layers.Dense(1, activation=kr.activations.relu))

model.compile(optimizer=kr.optimizers.Adam(), loss=kr.losses.huber)
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=200)
```

```python
import pandas as pd
```

```python
y_hat = model.predict(X_test)
```

```python
pd.DataFrame({'True': y_test,
             'Predicted': y_hat.reshape(-1)})
```

```python
y_hat = y_hat.reshape(-1)
```
