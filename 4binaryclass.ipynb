```python
from sklearn.datasets import load_breast_cancer
```

```python
from tensorflow import keras as kr
```

```python
cancer = load_breast_cancer()
```

```python
X = cancer.data
y = cancer.target
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)
```

```python
model = kr.models.Sequential()

model.add(kr.layers.Dense(X.shape[1], activation=kr.activations.relu))
model.add(kr.layers.Dense(50, activation=kr.activations.relu))
model.add(kr.layers.Dense(100, activation=kr.activations.relu))
model.add(kr.layers.Dense(50, activation=kr.activations.relu))
model.add(kr.layers.Dense(1, activation=kr.activations.sigmoid))

model.compile(optimizer=kr.optimizers.Adam(), loss=kr.losses.binary_crossentropy, metrics=[kr.metrics.binary_accuracy])
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=200)
```

```python
y_hat = model.predict(X_test)
```

```python
y_hat = y_hat.round().astype(int).reshape(-1)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_hat))
```
