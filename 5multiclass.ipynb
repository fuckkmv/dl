```python
from tensorflow import keras as kr
```

```python
from sklearn.datasets import load_wine
```

```python
lw = load_wine()
X = lw.data
y = lw.target
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)
```

```python
model = kr.models.Sequential()

model.add(kr.layers.Dense(X.shape[1], activation=kr.activations.relu))
model.add(kr.layers.Dense(50, activation=kr.activations.relu))
model.add(kr.layers.Dense(100, activation=kr.activations.relu))
model.add(kr.layers.Dense(50, activation=kr.activations.relu))
model.add(kr.layers.Dense(3, activation=kr.activations.softmax))

model.compile(optimizer=kr.optimizers.Adam(), loss=kr.losses.sparse_categorical_crossentropy, metrics=[kr.metrics.sparse_categorical_accuracy])
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=200)
```

```python
y_hat = model.predict(X_test)
```

```python
y_hat = y_hat.argmax(axis=1)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_hat))
```
